```python
def get_temperature(node_id):
    query='select timestamp,temp from light_temp where node_id=%d;' %node_id
    result=psql.frame_query(query,mysql_conn)  
    return result
```

```python
def pre_process_temperature(df):
    df.index=pd.to_datetime(df.timestamp*1e9)
    df=df.drop('timestamp',1)
    df=df['2012':]
    df=df[df.temp>65]
    df=df[df.temp<130]
    return df   
```

```python
def aggregations_temperature(df):
    df_downsampled=df.resample('1D',how='mean')
    df_downsampled['min']=df.resample('1D',how='min')
    df_downsampled['max']=df.resample('1D',how='max')
    df_downsampled.rename(columns={'temp': 'mean'}, inplace=True)
    df_downsampled_celsius=df_downsampled.apply(lambda x:(x-32)/1.8)
    return [df_downsampled,df_downsampled_celsius]
```

```python
for node_id in range(2,8):
    df=get_temperature(node_id)
    df=pre_process_temperature(df)
    [df_downsampled,df_downsampled_celsius]=aggregations_temperature(df)
    plt.figure()
    df_downsampled_celsius.plot(title='Temperature variation in celsius in room'+str(node_id))
    fig=plt.gcf()
    fig.set_size_inches(16,9)        
    plt.savefig('%s.png' %node_id,dpi=100,bbox_inches='tight')    
```
